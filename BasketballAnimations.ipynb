```python

df = pd.read_csv('playoffStats.csv')
df = df[['season', 'fg3a_per_g', 'pts_per_g', 'fta_per_g', 'efg_pct']]
df.dropna(inplace=True)
def plot_animated_histogram(data, x_column, title):
    fig = px.histogram(data, x=x_column, animation_frame="season",
                       range_x=[data[x_column].min(), data[x_column].max()],
                       nbins=30, title=title)
    fig.update_layout(yaxis_title='Count', xaxis_title=title)
    fig.show()

plot_animated_histogram(df, 'fg3a_per_g', '3PT Attempts per Game')
plot_animated_histogram(df, 'pts_per_g', 'Total Points per Game')
plot_animated_histogram(df, 'fta_per_g', 'Free Throw Attempts per Game')
plot_animated_histogram(df, 'efg_pct', 'Effective FG %')
```

```python
df = pd.read_csv('playoffStats.csv')
df = df[['season', 'fg3a_per_g', 'fg3_pct', 'ft_pct', 'fg3_per_g']]
df.dropna(inplace=True)

fig = px.scatter(
    df,
    x='fg3a_per_g',
    y='fg3_per_g',
    animation_frame='season',
    animation_group='fg3_per_g',
    size='ft_pct',
    color='fg3_pct',
    hover_name='fg3_per_g',
    range_x=[df['fg3a_per_g'].min(), df['fg3a_per_g'].max()],
    range_y=[df['fg3_per_g'].min(), df['fg3_per_g'].max()],
    color_continuous_scale=px.colors.sequential.Plasma,
    labels={'fg3a_per_g': '3PT Attempts per Game', 'fg3_per_g': '3PT Made per Game', 'ft_pct': 'Free Throw %', 'fg3_pct': '3PT %'}
)
fig.update_layout(title='NBA Playoff Player Performance Over Years')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('playoffStats.csv')
print(df[['season', 'fg3a_per_g', 'fg3_pct', 'ft_pct', 'fg3_per_g', 'pos']].isna().sum())
df = df[['season', 'fg3a_per_g', 'fg3_pct', 'ft_pct', 'fg3_per_g', 'pos']]
df.dropna(subset=['fg3a_per_g', 'fg3_pct', 'ft_pct', 'fg3_per_g', 'pos'], inplace=True)

fig = px.scatter(
    df,
    x='fg3a_per_g',
    y='fg3_pct',
    animation_frame='season',
    animation_group='fg3_pct',
    size='ft_pct',
    color='pos',
    hover_name='fg3_pct',
    range_x=[df['fg3a_per_g'].min(), df['fg3a_per_g'].max()],
    range_y=[df['fg3_pct'].min(), df['fg3_pct'].max()],
    color_discrete_map={'C': 'blue', 'PG': 'red', 'SG': 'green', 'SF': 'purple', 'PF': 'orange'},  # Custom colors for positions
    labels={'fg3a_per_g': '3PT Attempts per Game', 'fg3_per_g': '3PT%', 'ft_pct': 'Free Throw %', 'fg3_pct': '3PT %', 'pos': 'Position'}
)

fig.update_layout(title='NBA Playoff Player Performance Over Years by Position')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('playoffStats.csv')
df = df[['season', 'fg3a_per_g', 'fg3_pct', 'age', 'pos']]
df.dropna(subset=['fg3a_per_g', 'fg3_pct', 'age', 'pos'], inplace=True)
df.sort_values('season', ascending=True, inplace=True)
color_scale = [[0, 'green'], [1, 'red']]

fig = px.scatter(
    df,
    x='fg3a_per_g',
    y='fg3_pct',
    animation_frame='season',
    animation_group='fg3_pct',
    size='fg3_pct',
    color='age',
    hover_name='age',
    range_x=[df['fg3a_per_g'].min(), df['fg3a_per_g'].max()],
    range_y=[df['fg3_pct'].min(), df['fg3_pct'].max()],
    color_continuous_scale=color_scale,
    labels={'fg3a_per_g': '3PT Attempts per Game', 'fg3_pct': '3PT %', 'age': 'Player Age', 'pos': 'Position'}
)

fig.update_layout(
    title='NBA Playoff Player 3-Point Performance Over Years by Age',
    paper_bgcolor='rgba(245, 246, 250, 1)',
    plot_bgcolor='rgba(245, 246, 250, 1)',
    font=dict(family="Arial, sans-serif", size=12, color="RebeccaPurple")
)
fig.show()
```
